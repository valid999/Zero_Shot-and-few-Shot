How does Zero Shot Learning work ?
Zero Shot Learning is when we are testing a model on a task for which it hasn't been trained.


In Zero Shot Classification we are  asking the model to do classification for labels which modelk has noyseen during the model .

one of the popular methods for zero shot learning in natural language Inference NLI



# Task

Let's say we have the sentence **One day I will see the world** and we would like to classify whether this sentence is about

1. Travel
2. Cooking
3. dancing

Now for all three classification labels . we can have three hypothesis

HYpothesis 1: This text about travel
Hypothesis 2: This text about cooking
Hypothesis 3: This text is about dancing

```python
from transformers import pipeline

classifier = pipeline("zero-shot-classification" , model = 'facebook/bart-large-mnli')
```

```python
Sequence_to_classify = "one day I will see the world"
candidate_labels = ['travel' , 'cooking' , 'dancing']


classifier(Sequence_to_classify , candidate_labels)
```

# If more than one candidate label can be correct

```python
candidate_labels = ['travel' , 'cooking' , 'dancing' , 'exploration']
classifier(Sequence_to_classify , candidate_labels , multi_label=True)
```

```python
sequence_to_classify = "Donlad Trump will be next president"
candidate_labels = ['science' , 'politics' , 'history']


classifier(sequence_to_classify , candidate_labels)
```
